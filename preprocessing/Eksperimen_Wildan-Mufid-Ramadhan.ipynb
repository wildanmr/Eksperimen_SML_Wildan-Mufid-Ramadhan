```python
# Visualizations
for column in df.columns:
    plt.figure(figsize=(8, 5))
    if df[column].nunique() < 10:
        sns.countplot(x=column, data=df)
        plt.title(f'Distribution of {column}')
        plt.xticks(rotation=45)
    else:
        sns.histplot(df[column], kde=True)
        plt.title(f'Distribution of {column}')
    plt.tight_layout()
    plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(15, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

# **5. Data Preprocessing**

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
from sklearn.preprocessing import StandardScaler

# Make a copy of the original dataset
df_processed = df.copy()

print(f"Dataset shape before preprocessing: {df_processed.shape}")

# 1. Handle Missing Values (if any)
print(f"Missing values before: {df_processed.isnull().sum().sum()}")
df_processed.dropna(inplace=True)
print(f"Missing values after: {df_processed.isnull().sum().sum()}")

# 2. Handle Duplicate Data
print(f"Duplicates before: {df_processed.duplicated().sum()}")
df_processed.drop_duplicates(inplace=True)
print(f"Duplicates after: {df_processed.duplicated().sum()}")

print(f"Dataset shape after cleaning: {df_processed.shape}")
```

```python
# 3. Feature Scaling/Normalization
target_column = 'Diabetes_binary'

if target_column in df_processed.columns:
    # Separate features and target
    X = df_processed.drop(target_column, axis=1)
    y = df_processed[target_column]
    
    # Scale numerical features
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns, index=X.index)
    
    # Combine scaled features with target
    df_processed = pd.concat([X_scaled, y], axis=1)
    
    print("Features have been standardized.")
else:
    print(f"Target column '{target_column}' not found. Please adjust the target_column variable.")
    print(f"Available columns: {list(df_processed.columns)}")

# Display the processed data
print("\nProcessed dataset:")
df_processed.head()
```
